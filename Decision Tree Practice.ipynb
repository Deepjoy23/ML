## Overfit and Underfit

```python
# Generate a simple dataset
x, y = make_classification(n_samples=100, n_features=2, 
                           n_informative=2, n_redundant=0, 
                           random_state=1, n_clusters_per_class=1)
```

```python
# train and test spliting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
# DecisionTreeClassifier function using to decided Decesion Treee
model = DecisionTreeClassifier(criterion ="entropy",max_depth=20,min_samples_split=5)
```

```python
# Training and Training score fiting
model.fit(x_train,y_train)
```

```python
# predict x_train data
y_train_predict =model.predict(x_train)
y_train_predict
```

```python
# confusion matrix, classification_report, precision_score, recall_score, f1_score importing
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
```

```python
# f1_score of y_train and y_train_predict
f1_score_training = f1_score(y_train, y_train_predict)
f1_score_training
```

```python
# predict x_test data 
y_test_predict = model.predict(x_test)
y_test_predict
```

```python
# f1_score of y_test and y_test_predict
f1_score_testing = f1_score(y_test, y_test_predict)
f1_score_testing
```

## Regression

```python
# Regression Data Making
X = 6 * np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9*X + 2 + np.random.randn(200, 1)

plt.plot(X, y, 'b.')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# train and test spliting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
# importing Decision Tress Regression
from sklearn.tree import DecisionTreeRegressor
```

```python
model_reg = DecisionTreeRegressor()
model_reg.fit(x_train, y_train)
```

```python
# y_train_pred calculationg from x_train
y_train_pred = model_reg.predict(x_train)
y_train_pred
```

```python
y_train.reshape(1,-1)[0].shape
```

```python
y_train_pred.shape
```

```python
# R2 Score checking of y_train and y_train_pred
from sklearn.metrics import r2_score # "r2_score" imporing from "sklearn.metrics" folder
r2_score(y_train, y_train_pred)
```

```python
# R2 Score checking of y_test and y_test_predict
y_test_predict = model_reg.predict(x_test)
r2_score(y_test, y_test_predict)
```
