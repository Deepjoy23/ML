## Clustering

```python
import matplotlib.pyplot as plt

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.figure(figsize=(10,3))
plt.scatter(x, y)
plt.title("Scatter Distribution of Points")
plt.xlabel("X-axis",)
plt.ylabel("Y-axis")
plt.show()
```

```python
# Importing KMeans folder from sklearn.cluster
from sklearn.cluster import KMeans
```

```python
# Points X-Axis data
x
```

```python
# Points Y-Axis data
y
```

```python
data= list(zip(x,y))
data
```

```python
# Blank or null list creats
inertias =[]
inertias
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i) # KMeans trains from 1 to 10 clusters
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
```

```python
# inertias (centroids) are checking
inertias
```

```python
# Elbow Method Graph
plt.plot(range(1,11),inertias,marker="o")
plt.title("Elbow Method")
plt.xlabel("Cluster Number")
plt.ylabel("Inertias or Centroids")
plt.show()
```

```python
# Scatter Distribution of Points in Kmeans Clustering (Llyoid's Algorithm)
kmeans = KMeans(n_clusters = 2) # From Elbow Method Graph we check the Elbow at Cluster Number = 2
kmeans.fit(data)
plt.scatter(x, y, c= kmeans.labels_)
plt.title("Scatter Distribution of Points in Kmeans Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Lebeling of points in Kmeans Clustering (Llyoid's Algorithm)
kmeans.labels_
```

```python
# importing data ("Iris.csv")
import pandas as pd
data = pd.read_csv("Iris.csv")
data
```

```python
# Importing KMeans folder from sklearn.cluster
from sklearn.cluster import KMeans
```

```python
# Blank or null list creats
inertias =[]
inertias
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]])
    inertias.append(kmeans.inertia_)
```

```python
inertias
```

```python
# Elbow Method Graph
plt.plot(range(1,11),inertias,marker="o")
plt.title("Elbow Method")
plt.xlabel("Cluster Number")
plt.ylabel("Inertias or Centroids")
plt.show()
```

```python
# Fitting the KMeans Clustering (Llyoid's Algorithm)
kmeans = KMeans(n_clusters = 2) # From Elbow Method Graph we check the Elbow at Cluster Number = 2
kmeans.fit(data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]])
```

```python
# Lebeling of points in Kmeans Clustering (Llyoid's Algorithm)
kmeans.labels_
```

```python
# scatter plot in seaborn of ("Iris.csv")
import seaborn as sns

plt.figure(figsize=(5,5))
sns.scatterplot(data = data, x = "SepalLengthCm", y="PetalLengthCm", hue="Species") # hue use for filttering data with "Species" category 
plt.show()
```
